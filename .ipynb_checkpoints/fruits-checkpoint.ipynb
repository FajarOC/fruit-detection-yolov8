```python
from ultralytics import YOLO

# Inisialisasi model YOLOv8
model = YOLO('yolov8n.pt')  # Menggunakan versi YOLOv8 nano sebagai dasar
```

```python
model.train(data='c:/Users/fajar/fruitdeteksi/datasets/data.yaml', epochs=10) # atur parameter disini
```

```python
model.save("C:/Users/fajar/fruitdeteksi/fruit_detection_model.pt")
```

```python
import pandas as pd
from ultralytics import YOLO

# Perform model validation
metrics = model.val()

# Retrieve mean results (precision, recall, mAP@0.5, mAP@0.5-0.95)
mean_metrics = metrics.mean_results()

# Create a dictionary for the metrics
metrics_data = {
    'Metric': ['Precision (mp)', 'Recall (mr)', 'mAP@0.5', 'mAP@0.5:0.95'],
    'Value': [
        mean_metrics[0],  # Mean Precision
        mean_metrics[1],  # Mean Recall
        mean_metrics[2],  # mAP at IoU threshold of 0.5
        mean_metrics[3]   # mAP at IoU thresholds from 0.5 to 0.95
    ]
}

# Convert to DataFrame for clean table format
metrics_df = pd.DataFrame(metrics_data)

# Display the table
print(metrics_df)
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt

# Use %matplotlib inline to display plots directly in the notebook
%matplotlib inline

# Load model yang telah dilatih
model = YOLO("C:/Users/fajar/fruitdeteksi/fruit_detection_model.pt")

# Load gambar untuk diuji
image_path = "C:/Users/fajar/fruitdeteksi/images/fruit.jpeg"
image = cv2.imread(image_path)

# Convert the image from BGR to RGB (as OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Lakukan prediksi pada gambar
results = model(image_rgb)

# Jika hasil prediksi berupa list, ambil elemen pertama
if isinstance(results, list):
    results = results[0]

# Get the image with bounding boxes and labels from the results
annotated_img = results.plot()

# Plot original image and detection result side by side
plt.figure(figsize=(15, 7))

# Original image
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')  # Turn off the axis

# Detection result image
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.imshow(annotated_img)
plt.title('Detection Result')
plt.axis('off')  # Turn off the axis

# Show both images
plt.show()
```
